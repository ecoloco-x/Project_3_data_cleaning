```python
import pandas as pd
import numpy as np
```

```python
pollution_cleaning= pd.read_csv("pollution_uncleaned.csv")
pollution_cleaning
```

```python
# dropping "Unnamed: 0"
pollution_cleaning.drop(columns="Unnamed: 0", inplace=True)
pollution_cleaning
```

```python
# changing area to either "Coastal Waters" or "Port Limits"

pollution_cleaning["Area"].value_counts()
```

```python
pollution_cleaning["Area"] = pollution_cleaning["Area"].str.lower()
pollution_cleaning["Area"].value_counts()
```

```python
# Marina, Port, Port limits => Port Limits
pollution_cleaning.loc[pollution_cleaning["Area"] == "port", "Area"] = "port limits"
pollution_cleaning.loc[pollution_cleaning["Area"] == "marina", "Area"] = "port limits"
pollution_cleaning.loc[pollution_cleaning["Area"] == "port", "Area"] = "port limits"

# Coastal waters, Coastal, Territorial Sea, Territorial sea, Offshore => Coastal Wtarer
# GBRMP, GBR: means Great Barrier Reef Marine Park => Coastal Water
pollution_cleaning.loc[pollution_cleaning["Area"] == "coastal", "Area"] = "coastal waters"
pollution_cleaning.loc[pollution_cleaning["Area"] == "gbrmp", "Area"] = "coastal waters"
pollution_cleaning.loc[pollution_cleaning["Area"] == "gbr", "Area"] = "coastal waters"
pollution_cleaning.loc[pollution_cleaning["Area"] == "territorial sea", "Area"] = "coastal waters"
pollution_cleaning.loc[pollution_cleaning["Area"] == "offshore", "Area"] = "coastal waters"
pollution_cleaning["Area"].value_counts()

# check where to put Inland waters => Coastal? Port ?
```

```python
# checking what inland are with location description
inland_pollution = pollution_cleaning[pollution_cleaning['Area'] == "inland waters"]
inland_pollution
```

```python
# a Google maps search shows that the "inland waters" are all within marina or really close to a city with port so they can be coded as "port limits"
```

```python
# inland waters => Port Limits
pollution_cleaning.loc[pollution_cleaning["Area"] == "inland waters", "Area"] = "port limits"
pollution_cleaning["Area"].value_counts()
```

```python
def remove_trailing_spaces(x):
    for col in x.columns:
        x[col] = x[col].str.rstrip()
    return x
```

```python
pollution_cleaning["Pollutant"] = pollution_cleaning["Pollutant"].str.lower()
remove_trailing_spaces(pollution_cleaning)
```

```python
pollution_cleaning["Pollutant"].value_counts()
# 92 different types
```

```python
pollution_cleaning["Pollutant"] = pollution_cleaning["Pollutant"].str.replace('other - ','')
pollution_cleaning["Pollutant"] = pollution_cleaning["Pollutant"].str.replace("""other \(""",'')
pollution_cleaning["Pollutant"] = pollution_cleaning["Pollutant"].str.replace("""\)""",'')
pollution_cleaning["Pollutant"].value_counts()
```

```python
all_pollu = pd.DataFrame(pollution_cleaning["Pollutant"].value_counts())
```

```python
# uniformize lubricant oil
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "lube", "Pollutant"] = "lubricant oil"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "lube oil", "Pollutant"] = "lubricant oil"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "lubricating", "Pollutant"] = "lubricant oil"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "lubricating oil", "Pollutant"] = "lubricant oil"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "lubrication oil", "Pollutant"] = "lubricant oil"
pollution_cleaning["Pollutant"].value_counts()
```

```python
# uniformize hydraulic oil
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "hyd", "Pollutant"] = "hydraulic oil"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "hydraulic", "Pollutant"] = "hydraulic oil"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "hydraulic fluid", "Pollutant"] = "hydraulic oil"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == " hydraulic", "Pollutant"] = "hydraulic oil"
pollution_cleaning["Pollutant"].value_counts()
```

```python
# unifomizin diesel
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "diesel sheen", "Pollutant"] = "diesel"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "carbon from diesel engine", "Pollutant"] = "diesel"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "engine oil", "Pollutant"] = "diesel"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == " petrol", "Pollutant"] = "diesel"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "fuel oil", "Pollutant"] = "diesel"
```

```python
#unformizing bilge
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "bilge water", "Pollutant"] = "bilge"
```

```python
# others
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "heavy fuel oil", "Pollutant"] = "hfo"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "oil", "Pollutant"] = "oil (unspec)"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "petrol", "Pollutant"] = "oil (unspec)"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "grease trap", "Pollutant"] = "grease"
pollution_cleaning.loc[pollution_cleaning["Pollutant"] == "carbon soot", "Pollutant"] = "soot"
```
